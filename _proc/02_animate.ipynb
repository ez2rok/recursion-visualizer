# Fibonacci Example

```python
@Memoize()
def fib(n, edge_label=''):
  if n <= 2: return 1

  val = 0
  for decrement in [1, 2]:
    val += fib(n-decrement, edge_label='-{}'.format(decrement))
  return val
```

```python
#|output: true
fig, result = fib(5)
```
